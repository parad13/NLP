```python

grammer='NP:{<DT>?<JJ>*<NN>}' #determiner?adjective*common noun -> The studious boy

parser = nltk.RegexpParser(grammer)

output=parser.parse(tagged_words) #tagged words created in above step
print(output)

#for drawing sparse tree
# output.draw()
```

```python
#Chinking example:
# * - 0 or more repetations
# + - 1 or more repetations
# ? - optional character

#regex chinking(deciding grammer using regex)
#Grammer language/REGEX

grammer = r""" NP: {<.*>+}
            }<JJ>+{"""

#Creating parser:
parser = nltk.RegexpParser(grammer)

#parsing strinng :
output=parser.parse(tagged_words)
print(output)

#To visulize
# output.draw()
```

```python
#NER
#Sentence for Named Entity Recognition(REZ):
# nltk.download('maxent_ne_chunker')
# nltk.download('words')

sentence='Mr. smith made a deal on a beach of Switzerland near WHO'

#Tokenizing words :
tokenized_words = word_tokenize(sentence)

#PoS tagging :
for w in tokenized_words:
    tagged_words = nltk.pos_tag(w)
    
#NER :
N_E_R = nltk.ne_chunk(tagged_words,binary=True)

print(N_E_R)

# N_E_R.draw()
```

```python
#Sentence for NER:
sentence="Figure 100: Python implementation when a binary value is False."

#Tokenizing words :
tokenized_words = nltk.word_tokenize(sentence)

#PoS tagging :
for i in tokenized_words:
    tagged_words = nltk.pos_tag(i)
    
#NER :
N_E_R = nltk.ne_chunk(tagged_words,binary=True)

print(N_E_R)

# N_E_R.draw()
```

```python
#Sentence for NER:
sentence="When the binary value is True, then it will only show whether a particular entity is named entity or not. It will not show any further details on it."

#Tokenizing words :
tokenized_words = nltk.word_tokenize(sentence)

#PoS tagging :
for i in tokenized_words:
    tagged_words = nltk.pos_tag(i)
    
#NER :
N_E_R = nltk.ne_chunk(tagged_words,binary=False)

print(N_E_R)

# N_E_R.draw()
```

```python
#all details for all meanings of a word .
#word meaning with defination
for words in wordnet.synsets('Kind'):
    print(words.name(),'a')
    print(words.definition(),'b')
    print(words.examples(),'c')
    for lemma in words.lemmas():
        print(lemma,'d')
    print('\n')    #name a->def b ->ex c-> lemma d
```

```python
#Hypernyms :
word = wordnet.synsets('Power')[2]


#Find more abstract term
print(word.hypernyms(),'a')

#Find more specific term :
print(word.hyponyms(),'b')

#Get only name :
print(word.lemmas()[0].name(),'c')
print(word.lemmas()[1].name(),'d')
#print(word.lemmas()[2].name()) #list out of index
#print(word.lemmas()[3].name())
```

```python
#synonyms
synonyms=[]
for words in wordnet.synsets('Fun'):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())
        
synonyms
```

```python
#antonyms
antonyms=[]
for words in wordnet.synsets('HArd'):
    for lemma in words.lemmas():
        if lemma.antonyms():  #antonyms require if loop inside word.lemmas() for loop bcz lemmas aare already synonyms
            antonyms.append(lemma.antonyms()[0].name())
        
antonyms
```

```python
#synonyms & antonyms
synonyms=[]
antonyms=[]
for words in wordnet.synsets('Hard'):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())
        
print(synonyms)
print('\n')
print(antonyms)
```

## CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

#text for analysis
sentences = [
             "Jim and Pam traveled by bus.",
              "The train was late",
              "The flight was full. Traveling by flight is expensive."
]

#Create an object:
cv = CountVectorizer()

#Generating output for Bag of Words:
B_O_W=cv.fit_transform(sentences).toarray()

#Total words with their index in model:
print(cv.vocabulary_)
print('\n')

#Features :
print(cv.get_feature_names())
print('\n')

#show output :
print(B_O_W)
```

## TfidfVectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer

#Sentences for analysis :
sentences = ["This is the first document","This documnet is the second document"]

#Create an object:
vectorizer = TfidfVectorizer(norm = None)

#Generating output for TF_IDF :
X = vectorizer.fit_transform(sentences).toarray()

#Total words with their index in model:
print(vectorizer.vocabulary_)
print('\n')

#Features :
print(vectorizer.get_feature_names())
print('\n')

#show output :
print(X)
```
